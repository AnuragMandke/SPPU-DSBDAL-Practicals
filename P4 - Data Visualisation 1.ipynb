```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

```python
titanic.dtypes
```

<h2>1. Countplot</h2>

```python
# 1. Countplot: See survival based on gender
sns.countplot(x='survived', hue='sex', data=titanic)
plt.title('Survival Count by Gender')
plt.show()
print("\nObservation: Women had a better survival rate than men.\n")

# 2. Countplot: See survival based on Pclass (Ticket Class)
sns.countplot(x='survived', hue='pclass', data=titanic)
plt.title('Survival Count by Class')
plt.show()
print("\nObservation: Women had a better survival rate than men.\n")

#The countplot() function in Seaborn is used to create a bar plot that displays the count of categorical data.
```

<h2>2. Pairplot</h2>

```python
# 3. Pairplot: See relationships between numerical features (age, fare, pclass)
sns.pairplot(titanic.dropna(), hue='survived', vars=['age', 'fare', 'pclass'])
plt.title('Pairplot of Age, Fare, and Pclass')
plt.show()
print("Observation: Survival was higher for wealthy (high fare) and upper-class (1st class) passengers,\nwhile lower-class (3rd class) passengers had a much lower survival rate.")

#Plots pairwise relationships between the selected numerical variables in this case age fare and pclass
```

<h2>3. Lineplot</h2>

```python
# Line Plot for fare
sns.lineplot(x=titanic.index, y='fare', data=titanic)
plt.title('Line Plot of Fare vs Passenger Index')
plt.xlabel('Index')
plt.ylabel('Fare')
plt.show()
print("Observation: Higher fares were mostly paid by 1st-class passengers, while lower fares were common for 3rd-class passengers.")
```

<h2> 4. Violinplot</h2>

```python
sns.violinplot(x="pclass", y="age", data=titanic, hue="survived", split=True)
plt.title("Age Distribution by Passenger Class")
plt.show()

# Creates a violin plot, which combines a box plot with a KDE (density plot).
```

<h2>5. KDEplot</h2>

```python
# KDE Plot: Fare distribution
sns.kdeplot(titanic['fare'], shade=True, color='blue')
plt.title('KDE Plot of Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
print("Observation: The fare distribution is positively skewed, with a few passengers paying significantly higher fares (likely 1st-class tickets).")

# Creates a Kernel Density Estimation (KDE) plot, which is a smoothed version of the histogram.
```

<h2>6. Barplot</h2>

```python
# Bar Plot: Average Fare by Pclass
sns.barplot(x='pclass', y='fare', data=titanic)
plt.title('Bar Plot of Average Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Average Fare')
plt.show()
```

<h2>7. Histogram</h2>

```python
# Plotting the histogram of the 'fare' column
sns.histplot(titanic['fare'], kde=True, bins=20, color='blue')

# Adding title and labels
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
